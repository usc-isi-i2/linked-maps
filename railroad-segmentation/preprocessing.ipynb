# Exploring preprocessing for the maps

```python
import torch
import numpy as np
import cv2
import tifffile
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt
```

```python
# image = cv2.imread('/Users/Pratulya/Desktop/isi/research/segmentation/data/segmentation_data/CO_Louisville_1965/map_CO_Louisville_1965_degeo.png')
image = cv2.imread('/Users/Pratulya/Desktop/isi/research/segmentation/data/segmentation_data/CA_Tustin_1965/map_CA_Tustin_301027_1965_24000_geo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image.shape
```

```python
pil_image = TF.to_pil_image(image)
```

#### Normalizing w/o contrast

```python
image_tensor = TF.to_tensor(pil_image)
normalized1 = TF.normalize(image_tensor, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
```

#### Normalizing w/ contrast

```python
contrast_image = TF.adjust_contrast(pil_image, 2)
image_tensor = TF.to_tensor(contrast_image)
normalized2 = TF.normalize(image_tensor, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
```

```python
normalized1.shape, normalized2.shape
```

# Example 1

```python
patch = image[1600:2100, 9450:9950]
plt.figure()
plt.imshow(patch)
plt.show()
```

```python
normalized_patch1 = normalized1.permute(1,2,0).numpy()[1600:2100, 9450:9950]
plt.figure()
plt.imshow(normalized_patch1)
plt.show()
```

```python
normalized_patch2 = normalized2.permute(1,2,0).numpy()[1600:2100, 9450:9950]
plt.figure()
plt.imshow(normalized_patch2)
plt.show()
```
